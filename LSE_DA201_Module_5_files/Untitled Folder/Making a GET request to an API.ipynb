```python
# Import libraries.
import requests
import json
```

```python
# Identify the URL to connect to.
# Create a variable to store the URL.
def get_response(site):
    response = requests.get(url=site)
    return response

# Send connect request.
bitcoin_response = get_response('https://api.coindesk.com/v1/bpi/currentprice.json')
population_response = get_response('https://datausa.io/api/data?drilldowns=Nation&measures=Population')

# View answer from URL.
print(bitcoin_response)
print(population_response)
```

```python
# Write if and else statements to get information and status_code.
if bitcoin_response.status_code == 200:
    print(bitcoin_response.headers)
else:
    print(bitcoin_response.status_code)
    bitcoin_response.headers['Content-Encoding']
```

```python
# Write if and else statements to get information and status_code.
if population_response.status_code == 200:
    print(population_response.headers)
else:
    print(population_response.status_code)
    population_response.headers['Content-Encoding']
```

```python
# Change response to json.
bitcoin_response.json()
```

```python
# Change response to json.
population_response.json()
```

```python
# Parse JSON data with loads().
bitcoin_json = json.loads(bitcoin_response.text)

# View the content.
print(bitcoin_json)

# Formatting JSON.
print(json.dumps(bitcoin_json, indent=4))
```

```python
# Parse JSON data with loads().
population_json = json.loads(population_response.text)

# View the content.
print(population_json)

# Formatting JSON.
print(json.dumps(population_json, indent=4))
```

```python
import pandas as pd 

# Create a DataFrame directly from the output.
bitcoin_df = pd.DataFrame(bitcoin_json)

# View the DataFrame.
bitcoin_df.head()
```

```python
# Save the JSON file to .json.
# Create a JSON file.
bitcoin_file = json.dumps(bitcoin_json)


with open('bitcoin_file.json', 'w') as f:
    json.dump(bitcoin_json, f)
    
# Save as a CSV file without index.
bitcoin_df.to_csv('bitcoin_csv.csv', index=False)
```
