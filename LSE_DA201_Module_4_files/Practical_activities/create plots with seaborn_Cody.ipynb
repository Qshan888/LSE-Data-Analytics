```python
# Import Matplotlib, Seaborn, and Pandas.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Import the csv file.
movies = pd.read_csv('../Data/movies.csv')

# View the DataFrame.
print(movies.shape)
print(movies.dtypes)
print(movies.columns)
movies.head()
```

```python
# Import the csv file.
ott = pd.read_excel('../Data/ott.xlsx')

# View the DataFrame.
print(ott.shape)
print(ott.dtypes)
print(ott.columns)
ott.head()
```

```python
mov_ott = movies.merge(ott, how='left', on='ID')

print(mov_ott.shape)
mov_ott.head()
```

```python
# Plot a countplot to determine what age group has streamed the most movies.
sns.countplot(x='Age', data=mov_ott)
```

```python
# Create a countplot based on number of movies streamed by Netflix per age group.
sns.countplot(x='Age', hue='Netflix', data=mov_ott)
```

```python
# Create a histogram based on IMDb. 
sns.histplot(data=mov_ott, x='IMDb',binwidth=1)
```

```python
# Create a scatterplot with two variables (IMDb and Rotten Tomatoes).
sns.scatterplot(x='IMDb', y='Rotten Tomatoes', data=mov_ott)
```
